```python

# working with dataset name
splitname = dfname.split(".")
dname = str(splitname[0])
namel = dname.split("_")
dname1 = namel[0]
dnum1 = namel[1]
print("DATAFRAME ----------------- : ", dfname, dname, dname1, dnum1, driverPath)

# reading dataset from folder
aname = "../target/target_"+str(dnum1)+".csv"
df = pd.read_csv(aname)
try:
    df = df.drop(['Unnamed: 0'], axis = 1) 
except:
    print("no unnamed in submitform")
d = list(df["url"])
print()
print("Running runtarget.py in :", servername, jobname, dfname)
print()


# querying submitform data from mongodb database and instantiating the constants here 
submitformdatalist = list(datacol.find({}))
print("Submitform data list is :", submitformdatalist)

for sfEach in submitformdatalist:
    camp = sfEach['VERSION']
    if camp == submitFormVersion:
        NAME = sfEach['USERNAME']
        FIRSTNAME = sfEach['FIRSTNAME']
        LASTNAME = sfEach['LASTNAME']
        TITLE = sfEach['TITLE']
        EMAIL = sfEach['EMAIL']
        SUBJECT = sfEach['SUBJECT']
        MESSAGE = sfEach['MESSAGE']
        OPTION1 = sfEach['OPTION1']

        BRAND = sfEach['BRAND']
        POSITION = sfEach['DESIGNATION']
        WEBSITE = sfEach['WEBSITE']
        AGE = sfEach['AGE']
        NUMBER = sfEach['MOBILE_TELEPHONE_NUMBER']
        
        CITY = sfEach['CITY']
        POSTCODE = sfEach['POSTCODE']
        STATE = sfEach['STATE']
        CONTINENT = sfEach['CONTINENT']
        COUNTRY = sfEach['COUNTRY']
        
         
print()
print("HERE IS ALL THE CAMPAIGN DATA : ", camp, NAME, FIRSTNAME, LASTNAME, TITLE, POSITION,  EMAIL, SUBJECT, MESSAGE, OPTION1, BRAND, WEBSITE)
print()
print()
print()
        

#all the lists
namelist = ['name', 'Name',"Your", "your","User","user", "username","Username", "yourname", "Yourname","user name","User name", "name user","name User", "your name", "Your name", "user_name","User_name","name_user","name_User", "your_name", "Your_name","user-name","User-name", "your-name", "Your-name"]
firstnamelist = ["firstname", "FirstName","firstName", "Firstname","primary", "Primary","first name", "First Name","first Name", "First name", "first_name", "First_Name","first_Name", "First_name","first-name", "First-Name","first-Name", "First-name","first","First", "initial","Initial", "namefirst", "NameFirst","Namefirst", "nameFirst","name first", "Name First","Name first", "name first","name_first", "Name_First","Name_first", "name_First" ]
lastnamelist = ["lastname", "LastName","lastName", "Lastname","secondary", "Secondary","last name", "Last Name","last Name", "Last name", "last_name", "Last_Name","last_Name", "Last_name","last-name", "Last-Name","last-Name", "Last-name","last","Last", "namelast", "NameLast","Namelast", "nameLast","name last", "Name Last","Name last", "name last","name_last", "Name_Last","Name_last", "name_Last" ]
titlelist = ["title", "Title", "Mr","mr","Mr.","Mrs","mrs","mrs.","Mrs.","Ms","ms","Ms.","ms."]
subjectlist = ['Subject', "subject", 'topic', "topic", "what", "What", "Know", 'know']
numberlist = ['number', 'digit','digits',"Phone","phone","cell","Cell", "integer", "Integer", "Digits","Digit","Number","mobile","Mobile","telephone","Telephone","tele","Tele" ]
emaillist = ['email', 'Email', "mail", "Mail", "emailaddress","email_address","email-address","EmailAddress"]
brandlist = ['compnay', 'brand', 'business', 'official', 'corporate', 'startup', 'initiative']
positionlist = ["Job",'job', 'designation', "Designation","position", "Position", "role","Profession", "role","profession", "Responsibility","responsibility", ]
agelist = ['age', "Age","old","Old","year", "Year", "years","Years"]
websitelist = ['site', 'website','Site', 'Website', 'web', "Web", 'url','Url' ]
citylist = ['city', "City"]
stateprovincelist = ["state","State","States", "states", "Division","division"]
continentlist = ["Region","region",'continent',"Continent"]
countrylist = ['country', "Country"]
zippostlist = ['zip', "post","code",'Zip', "Post","Code",]
optionlist = ['optional', "Optional","addition","additional","more", "More", "Extra","extra"]
messagelist = ['text', "Text","message","msg","description", "Description","Information", "information","Message","question","Question","comment", "Commnet", "Note", "note" "something","Something", "Inquiry", "inquiry"]

# urllist = ['site', 'website','Site', 'Website', "Web", "web", "url","Url" ]
# locationlist = ['location', "address","Location","Address",'place',"Place"]
# successlist = ["thank", "thank you", "thanks",'Thanks', "Thank you","Thank","Welcome","welcome", "Hello", "Good", "hello","congratulations", 'congratulation', "Congratulations","Congratulation"]
values = [NAME, FIRSTNAME, LASTNAME, SUBJECT, TITLE, EMAIL, MESSAGE, AGE, WEBSITE, POSITION, BRAND, NUMBER, STATE, CITY, CONTINENT, COUNTRY, POSTCODE, OPTION1 ]
lists = [namelist, firstnamelist, lastnamelist, subjectlist, titlelist, emaillist, messagelist, agelist, websitelist, positionlist, brandlist, numberlist, stateprovincelist, citylist, continentlist, countrylist, zippostlist, optionlist]

    
def setvalue(values, lists,name, placeholder, idd, data, eachinput, inputted_something):
    count = 0
    for ii in range(len(lists)):
        if count == 0:
            for e in lists[ii]:
                if len(data) != 0:
                    try:
                        if e in data[idd]:
                            if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                                eachinput.send_keys(values[ii]);
                                count+=1
                                inputted_something+=1
                                break
                    except:
#                         print("There is a problem m sure")
                        continue
                        
                else:
                    if e in name or e in placeholder or e in idd:
                        if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                            eachinput.send_keys(values[ii]);
                            count+=1
                            inputted_something+=1
                            break          
    if count == 0:
        if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
            eachinput.send_keys(values[0])
            inputted_something+=1
    return inputted_something

    
######################################################################
def textareainput(each, inputted_something):
    #print("inputted_something before textarea ", inputted_something)
    c= True
    #workng with text input
    textareas = each.find_elements(By.TAG_NAME, 'textarea')
    for eachtarea in textareas:
        if c:
            eachtarea.send_keys(MESSAGE); 
            c = False
            inputted_something+=1
            #print("inputted_something after ta ", inputted_something)
    return inputted_something

######################################################################

def inputsection(data, total, each, inputted_something,  clicked, form_inputted):
    inputs = each.find_elements(By.TAG_NAME, 'input')
#     print("inside input")
    
    #working with each input
    for ei in range(len(inputs)):
        eachinput = inputs[ei]
        typee  = eachinput.get_attribute("type")
        idd  = eachinput.get_attribute("id")
        name  = eachinput.get_attribute("name")
        placeholder  = eachinput.get_attribute("placeholder")

        #Hidden
        if typee == "hidden":
#             print("Hidden")
            continue

        #Password
        elif typee == "password":
            #print("password based form is not acceptable")
            break

        #Text
        elif typee == "text":
            #print("inputted_something before setvalue ", inputted_something)
            #setting the value for iteration throught one input when type is text
            inputted_something = setvalue(values, lists,name, placeholder,idd, data, eachinput,inputted_something)
            #print("inputted_something after setvalue ", inputted_something)
            
        #other type of data such as email           
        elif typee == "email":
            #print("inputted_something before email ", inputted_something)
            if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                eachinput.send_keys(EMAIL)
                inputted_something+=1
                #print("inputted_something after email ", inputted_something)

        #number
        elif typee == "number":
            for k in agelist:
                if k in idd or k in name or k in placeholder:
                    if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                        eachinput.send_keys(int(NUMBER));
                        inputted_something+=1
                        break
            for kk in numberlist:
                if kk in idd or kk in name or kk in placeholder:
                    if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                        eachinput.send_keys(int(NUMBER));
                        inputted_something+=1
                        break
            
        #url
        elif typee == "url":
            if eachinput.get_attribute('value') == None or eachinput.get_attribute('value') == "":
                eachinput.send_keys(URL)
                inputted_something+=1

        #radio
        elif typee == "radio":
            eachinput.click()
            inputted_something+=1      
       
        #checkbox
        elif typee == "checkbox":
            eachinput.click()
            inputted_something+=1
            
    #inputing text area content
    inputted_something = textareainput(each, inputted_something) 
    
    try:
        selects = each.find_elements(By.TAG_NAME, 'select')
#         print("inside select")
        for s in range(len(selects)):
            eachselect = selects[s]
            typee  = eachselect.get_attribute("type")
            #select
            if typee == "select":
                select = Select(eachselect)
                # select by visible text
                select.select_by_index(1)
                inputted_something+=1
            continue

    except:
        print("error in select")  
        
    #working with each input
    for el in range(len(inputs)):
        eachinput = inputs[el]
        typee  = eachinput.get_attribute("type")
        if typee == "submit":
            #print("submit type")
            if inputted_something >=2: 
                #print("Button in submit type")
                eachinput.click()
                form_inputted+=1
                total+=1
                clicked = True
                return total, form_inputted
            else:
                break
        
        # button
        elif typee == "button":
            #print("button type")
            if inputted_something>= 2:
                #print("Button in button type")
                eachinput.click()
                form_inputted+=1
                total+=1
                clicked = True
                return total, form_inputted
            else:
                break
    return total, form_inputted, inputted_something      
       
            
######################################################################
def checkbtn(hbtn, total, form_inputted):
    typee  = hbtn.get_attribute("type")
    idd  = hbtn.get_attribute("id")
    name  = hbtn.get_attribute("name")
    text = hbtn.text
    print("inside check button")
    clicked_Once = 0
    if clicked_Once == 0:
        for e in ['Submit','submit','SUBMIT', 'Send','send',"SEND","CONTACT", 'contact', 'Contact','MESSAGE','message','Message','reach', 'REACH', 'Reach', 'TEXT', 'text', 'Text']:
            if typee == 'submit':
                hbtn.submit()
                total+=1
                form_inputted+=1
                clicked_Once+=1
                #print("Submission successful yahooooo type", total)
                return total, form_inputted
            elif typee == 'button':
                hbtn.submit()
                total+=1
                form_inputted+=1
                clicked_Once+=1
                #print("Submission successful yahooooo type", total)
                

            elif e in idd:
                hbtn.submit()
                #print("Submission successful  id")
                total+=1
                form_inputted+=1
                clicked_Once+=1
                return total, form_inputted
            elif e in name:
                hbtn.submit()
                #print("Submission name")
                total+=1
                form_inputted+=1
                clicked_Once+=1
                return total, form_inputted
            elif e in text:
                hbtn.submit()
                #print("Submission name")
                total+=1
                form_inputted+=1
                clicked_Once+=1
                return total, form_inputted
            else:
                print("Button not found")
      
    return total, form_inputted

################################################################################################
def clickbutton(each, total, form_inputted, soup):
    #print("Inside click button")
    val = 0
    if val == 0:
        try:
            btns = each.find_elements(By.TAG_NAME, "button")
#             print("Form div button trying to click")
            for b in range(len(btns)):
                eachbtn = btns[b]
                total, form_inputted = checkbtn(eachbtn, total, form_inputted)
                return total, form_inputted 
        except:
            print("ERROR in btn group")
    else:
        try:
#             print("No button in form div")
            gbtn = soup.find_all('button')
            for g in range(len(gbtn)):
                eachgbtn = gbtn[g]
                total, form_inputted = checkbtn(eachgbtn, total, form_inputted)
                return total, form_inputted 
        except:
            print("ERROR in gbtn group")

    
        
        
################################################
def startworks(q, w, tlinks,  sent, not_sent, total):
    dff = tlinks[q:w]
    ldff = len(dff)
    
    for i in range(ldff):
        inputted_something = 0
        clicked = False
        form_inputted = 0
        ever_inputted = 0
        fi = 0

        try:
            url = dff[i]
            url1 = url +"/?utm_source=ForeMedia.net&utm_medium=Get-Paid-For-Impressions"
            print(url1)
            driver = webdriver.Chrome(driverPath, chrome_options=chrome_options)
            driver.get(url1)
            time.sleep(3)
            source = driver.page_source
            soup = BeautifulSoup(source, 'html.parser')
            forms = driver.find_elements_by_tag_name('form')
        except:
            print("Error during loading. . .")
            print("Continue to next website")
            not_sent.append(url)
            driver.quit()
            continue

        try: 
            #if no error in geting the page then move forward to the find form section
            for each in forms:
                print("form ", fi)
                fi+=1
                if form_inputted == 0:
                    data = []
                    labels = each.find_elements(By.TAG_NAME, 'label')

                    #if lebal is not empty
                    if labels:
                        try:
                            for lab in labels:
                                labelfor  = lab.get_attribute("for")
                                labeltext  = lab.text
                                data.append((labelfor, labeltext))
                            data = dict(data)
#                             print("Inside label")
#                             print(data)

                            #code wraps up
                            print("Going inside input section")
                            total, form_inputted, inputted_something = inputsection(data, total, each, inputted_something, clicked, form_inputted)


                        except:
                            try:
                                print("Error in label so trying without data item")
                                data = []
                                total, form_inputted, inputted_something = inputsection(data, total, each, inputted_something, clicked, form_inputted)
                            except:
                                continue
                    else:
                        #code wraps up
                        try:
                            print('else yo')

                            data = []
                            total, form_inputted, inputted_something = inputsection(data, total, each, inputted_something, clicked, form_inputted)
                        except:
                            print("Error inside else label section")
                            continue

                    if inputted_something >=2 and form_inputted==0:
                        print("lets submit the form")
                        try:
                            total, form_inputted = clickbutton(each, total, form_inputted , soup)
                            #print("Total ok form_inputted inside ", total, form_inputted)
                        except:
                            print("Error while called form submission. total is : ", total)
                else:
                    ever_inputted += 1

        except:
            print("Error in whole form div")
            not_sent.append(url)
            driver.quit()
            continue

        try:
            if form_inputted != 0:
#                 print("lets insert the link inside sent list ")
                sent.append(url)

#                 print("Ever value", ever_inputted)
#                 print('printing success message')
#                 time.sleep(2)
                print("Now Total is ", total)

                do = dff[i]
                sign = [".", ",","//","/","?","https","http","www",":","%"]
                for f in sign:
                    if f == '.':
                        do = do.replace(f, ".")
                    elif f == "http" or f == "https":
                        do = do.replace(f, "")
                    elif f == "/":
                        do = do.replace(f, "-")
                    else:
                        do = do.replace(f, "")
                        
                name = "../ss/ss_"+str(dnum1)+"/"+str(do)+".png" 
                driver.save_screenshot(name)

            if form_inputted == 0 and ever_inputted == 0:
                print("No form ever inputted in this website :", url)
                not_sent.append(url)

#             print("Going next website")
#             print()
            driver.quit()
        except:
            driver.quit()
            continue
    return sent, not_sent, total
            
############################## ##############################     
############################## ##############################         

def threadWork(a, b , d, cou, lock):
    total = 0
    sent = []
    not_sent = []

    tlinks = d[a:b]
    print("Tlink :",len(tlinks))
    l = len(tlinks)
    vv = []
    num_thread = 10
    stp = math.ceil(l/num_thread)
    for ip in range(0,l,stp):
        vv.append(ip)
    vv.append(l)
    print("LL is :",vv)
    
        
    threads = []
    for e in range(num_thread):
        t = threading.Thread(target=startworks, args=(vv[e],vv[e+1], tlinks,  sent, not_sent, total)) 
        threads.append(t)
        t.start()
    for tr in threads:
        tr.join()             
#     except:
#         print("Error in thread")

    
    lock.acquire()
    sname = "../sent/sent_"+str(dnum1)+".csv"
    nsname = "../notsent/notsent_"+str(dnum1)+".csv"
    
    s1 = pd.read_csv(sname)
    ns1 = pd.read_csv(nsname)
    try:
        s1 = s1.drop("Unnamed: 0", axis = 1)
        s1.columns = ["url"]
        ns1 = ns1.drop("Unnamed: 0", axis = 1)
        ns1.columns = ["url"]
    except:
        print("No unname")

    sentdf = pd.DataFrame(sent)
    sentdf.columns = ["url"]
    not_sentdf = pd.DataFrame(not_sent)
    not_sentdf.columns = ["url"]
    print("SENT DF ", sentdf)
    print("NOT SENT DF ", not_sentdf)

    frames = [s1, sentdf]
    frames1 = [ns1, not_sentdf]

    sdf = pd.concat(frames, ignore_index =True )
    nsdf = pd.concat(frames1, ignore_index =True )

    sdf.to_csv(sname)
    nsdf.to_csv(nsname)
    
    print("Total is :",total)
    print(cou,len(sent),"SENT ************************************* :", sent)
    print(cou,len(not_sent),"Not SENT ***************************************:", not_sent)
    lock.release()
                    
    
```
